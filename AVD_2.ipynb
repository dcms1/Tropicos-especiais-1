#2 AVD TROPICOS ESPECIAIS

```python
# Importando as Bibliotecas
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import dlib
```

```python
# Reconhecendo a imagem
imagem = cv2.imread('sorrir.jpg') 
# Carregando o detector de marcos faciais da dlib:
detector_pontos = dlib.shape_predictor('/content/shape_predictor_68_face_landmarks.dat')
face_detector = dlib.get_frontal_face_detector()
```

```python
#reconhecendo as faces
deteccoes = face_detector(imagem, 1)
# Para cada face detectada realizaremos as seguintes tarefas:
for face in deteccoes:
  #Encontramos os pontos para determonar com exatidao onde Ã© a face da pessoa
  pontos = detector_pontos(imagem, face)
  
  for ponto in pontos.parts():
    #marcamos os marcos faciais
    cv2.circle(imagem, (ponto.x, ponto.y), 2, (0,255,0), 1)
  # pegamos as cordenadas das faces
  l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
  # Desenhamos um quadrado sobre a face da pessoa
  cv2.rectangle(imagem, (l,t), (r, b), (0,255,255), 2)

#Mostra a imagem
cv2_imshow(imagem)
```
